# 0. Czym jest analiza danych

### Czym jest analiza danych?

Analiza danych to proces przeksztacania, oczyszczania i modelowania danych w celu odkrycia uytecznych informacji, wycignicia wnioskw oraz wsparcia procesw decyzyjnych. W kontekcie szeroko rozumianej nauki o danych, analiza danych stanowi kluczowy element, ktry czy rne metody i techniki wykorzystywane do pracy z danymi.

### Gwne etapy analizy danych

1. **Zbieranie danych**:
   - Gromadzenie danych z rnych rde, takich jak bazy danych, arkusze kalkulacyjne, pliki CSV, API, czujniki IoT, media spoecznociowe, itd.
   
2. **Przygotowanie danych**:
   - Oczyszczanie danych (usuwanie lub korygowanie bdw, brakujcych wartoci).
   - Transformacja danych (normalizacja, standaryzacja, agregacja).
   - Wstpna analiza eksploracyjna (EDA) w celu zrozumienia podstawowych cech danych.
   
3. **Eksploracja danych (Exploratory Data Analysis - EDA)**:
   - Wizualizacja danych (wykresy, histogramy, diagramy pudekowe).
   - Statystyki opisowe (rednia, mediana, odchylenie standardowe).
   - Wykrywanie wzorcw i anomalii.
   
4. **Prezentacja wynikw**:
   - Tworzenie raportw i dashboardw.
   - Interpretacja wynikw i wyciganie wnioskw.
   - Prezentacja wynikw interesariuszom.

### Techniki i narzdzia

Analiza danych wykorzystuje rnorodne techniki i narzdzia, ktre pomagaj w przeksztacaniu surowych danych w wartociowe informacje.

#### Techniki analizy danych

- **Statystyka opisowa**: Podstawowe miary statystyczne, takie jak rednia, mediana, moda, odchylenie standardowe.
- **Wizualizacja danych**: Tworzenie wykresw i diagramw w celu atwiejszego zrozumienia danych.
- **Modelowanie predykcyjne**: Techniki takie jak regresja, drzewa decyzyjne, lasy losowe, sieci neuronowe.
- **Analiza klasteryzacyjna**: Grupowanie podobnych obiektw w klastry.
- **Analiza koszykowa**: Technika uywana w analizie zachowa zakupowych klientw (np. algorytm Apriori).
- **Analiza szeregw czasowych**: Techniki analizy danych zmieniajcych si w czasie (np. ARIMA, LSTM).

#### Narzdzia analizy danych

- **Jupyter Notebook**: Interaktywne rodowisko do kodowania w Pythonie, idealne do eksploracji danych i tworzenia raportw.
- **Pandas**: Biblioteka Pythona do manipulacji i analizy danych strukturalnych.
- **NumPy**: Biblioteka do oblicze numerycznych w Pythonie.
- **Matplotlib i Seaborn**: Biblioteki do tworzenia wykresw i wizualizacji danych w Pythonie.
- **Scikit-learn**: Biblioteka do uczenia maszynowego w Pythonie.
- **SQL**: Jzyk zapyta uywany do zarzdzania i manipulacji danymi w bazach danych.

### Zastosowania analizy danych

Analiza danych znajduje zastosowanie w wielu dziedzinach, w tym:

- **Biznes**: Analiza zachowa klientw, prognozowanie sprzeday, optymalizacja cen, analiza ryzyka.
- **Finanse**: Analiza portfeli inwestycyjnych, prognozowanie kursw akcji, wykrywanie oszustw.
- **Medycyna**: Analiza wynikw bada klinicznych, diagnoza chorb, personalizowana medycyna.
- **Marketing**: Segmentacja klientw, analiza kampanii marketingowych, targetowanie reklam.
- **Sport**: Analiza wynikw sportowych, optymalizacja treningw, strategia gry.
- **Rzd**: Analiza danych demograficznych, prognozowanie ekonomiczne, zarzdzanie kryzysowe.

# Interaktywna praca z kodem i danymi

### Interaktywna praca z kodem i danymi

Interaktywna praca z kodem i danymi jest jednym z najwaniejszych aspektw wspczesnej analizy danych, programowania i uczenia maszynowego. Jupyter Notebook, jedno z najpopularniejszych narzdzi w tej dziedzinie, umoliwia uytkownikom interaktywn prac z kodem i danymi, co przynosi szereg korzyci zarwno dla pocztkujcych, jak i zaawansowanych uytkownikw.

#### Podstawy Jupyter Notebook

Jupyter Notebook to otwarte rodowisko, ktre pozwala na tworzenie i udostpnianie dokumentw zawierajcych kod, tekst, wykresy i inne elementy multimedialne. Notatnik skada si z komrek, ktre mog zawiera kod, tekst sformatowany za pomoc Markdown, rwnania LaTeX i wiele innych. Interaktywno Jupyter Notebook polega na moliwoci uruchamiania poszczeglnych komrek kodu niezalenie od siebie, co umoliwia iteracyjn prac i szybkie testowanie kodu.

#### Korzyci z interaktywnej pracy

1. **Natychmiastowa informacja zwrotna:** Jedn z najwikszych zalet Jupyter Notebook jest moliwo uzyskiwania natychmiastowej informacji zwrotnej. Uytkownicy mog uruchamia poszczeglne komrki kodu i od razu widzie wyniki, co uatwia debugowanie i eksperymentowanie z kodem.

2. **Eksploracja danych:** Interaktywne rodowisko Jupyter Notebook jest idealne do eksploracyjnej analizy danych (EDA). Uytkownicy mog atwo wczytywa, przeglda i przetwarza dane, a take tworzy wykresy i wizualizacje, ktre pomagaj zrozumie struktury i wzorce w danych.

3. **Dokumentacja i prezentacja:** Dziki komrkom tekstowym sformatowanym za pomoc Markdown, Jupyter Notebook umoliwia tworzenie dobrze udokumentowanych i przejrzystych notatnikw. Uytkownicy mog dodawa opisy, rwnania matematyczne, listy, linki i obrazy, co czyni notatniki wietnym narzdziem do prezentacji wynikw i wsppracy z innymi.

4. **Reprodukowalno:** Notatniki Jupyter pozwalaj na tworzenie kompletnych, samodzielnych dokumentw, ktre zawieraj cay kod, dane wejciowe i wyniki analiz. Dziki temu inni uytkownicy mog atwo odtworzy proces analizy, co jest kluczowe w badaniach naukowych i projektach zespoowych.

#### Typowe zastosowania

1. **Analiza danych:** Jupyter Notebook jest szeroko stosowany w analizie danych. Uytkownicy mog wczytywa dane z rnych rde (CSV, Excel, SQL, itp.), przetwarza je za pomoc Pandas, wykonywa obliczenia numeryczne z NumPy i tworzy zaawansowane wizualizacje za pomoc Matplotlib i Seaborn.

2. **Uczenie maszynowe:** Jupyter Notebook jest idealnym narzdziem do rozwijania i trenowania modeli uczenia maszynowego. Biblioteki takie jak Scikit-learn, TensorFlow i PyTorch s w peni kompatybilne z Jupyter, co umoliwia atwe eksperymentowanie z rnymi modelami i hiperparametrami.

3. **Nauczanie i edukacja:** Ze wzgldu na swoj przejrzysto i interaktywno, Jupyter Notebook jest czsto wykorzystywany w edukacji. Umoliwia tworzenie interaktywnych materiaw dydaktycznych, ktre uczniowie mog atwo zrozumie i z nimi pracowa.

### **Komrki kodu i komrki Markdown**
- **Komrki kodu:** Umoliwiaj pisanie i wykonywanie kodu Python bezporednio w notatniku.
- **Komrki Markdown:** Umoliwiaj dodawanie sformatowanego tekstu, nagwkw, list, linkw, obrazkw oraz formu LaTeX. Jest to przydatne do tworzenia dokumentacji i notatek.


### **Interaktywne uruchamianie kodu**
- **Shift + Enter:** Uruchamia biec komrk i przechodzi do nastpnej.
- **Ctrl + Enter:** Uruchamia biec komrk bez przechodzenia do nastpnej.
- **Alt + Enter:** Uruchamia biec komrk i wstawia now komrk poniej.

### **Autouzupenianie kodu i podpowiedzi**
- **Tab:** Autouzupenianie kodu, np. nazw zmiennych, funkcji, moduw.
- **Shift + Tab:** Wywietlanie podpowiedzi dla funkcji/metody, co jest szczeglnie przydatne do szybkiego przegldu dostpnych argumentw i ich opisw.

W Jupyter Notebook istnieje wiele przydatnych skrtw klawiszowych, ktre mog znacznie przyspieszy prac i poprawi produktywno. Skrty klawiszowe s podzielone na dwa tryby: tryb komend (Command Mode) i tryb edycji (Edit Mode).

### Tryb Komend (Command Mode)

W trybie komend (aktywowany klawiszem `Esc`), uytkownik moe wykonywa operacje na caych komrkach. Oto niektre z najwaniejszych skrtw:

- **Enter**: przejcie do trybu edycji
- **Shift + Enter**: uruchom biec komrk i przejd do nastpnej
- **Ctrl + Enter**: uruchom biec komrk
- **Alt + Enter**: uruchom biec komrk i wstaw now poniej
- **Y**: zmie komrk na kod (Code)
- **M**: zmie komrk na Markdown
- **A**: wstaw komrk powyej (Above)
- **B**: wstaw komrk poniej (Below)
- **D, D**: usu biec komrk
- **Z**: cofnicie usunicia komrki
- **C**: kopiuj biec komrk
- **V**: wklej skopiowan komrk poniej
- **Shift + V**: wklej skopiowan komrk powyej
- **X**: wytnij biec komrk
- **Shift + M**: scal zaznaczone komrki
- **H**: wywietl wszystkie skrty klawiszowe (pomoc)

### Tryb Edycji (Edit Mode)

W trybie edycji (aktywowany klawiszem `Enter`), uytkownik moe edytowa zawarto komrki. Oto kilka przydatnych skrtw:

- **Esc**: przejcie do trybu komend
- **Ctrl + Enter**: uruchom biec komrk
- **Shift + Enter**: uruchom biec komrk i przejd do nastpnej
- **Tab**: autouzupenianie kodu lub wcicie
- **Shift + Tab**: wywietlenie podpowiedzi dla biecej funkcji/metody
- **Ctrl + Shift + -**: podziel komrk w miejscu kursora
- **Ctrl + /**: komentuj/odkomentuj zaznaczony kod

### Skrty dotyczce formatowania Markdown

W trybie edycji Markdown, uytkownik moe korzysta z nastpujcych skrtw:

- **Ctrl + 1** do **Ctrl + 6**: zmie poziom nagwka (od H1 do H6)
- **Ctrl + B**: pogrubienie tekstu
- **Ctrl + I**: kursywa
- **Ctrl + K**: dodaj link

### Inne przydatne skrty

- **Ctrl + S**: zapisz biecy notatnik
- **Ctrl + F**: znajd w notatniku
- **Ctrl + Shift + P**: wywietl palet polece (umoliwia wyszukiwanie i wykonywanie rnych polece)

# Google Colab do pracy w chmurze

**Google Colab** to darmowa usuga stworzona przez Google, ktra umoliwia pisanie i uruchamianie kodu Python bezporednio w przegldarce, korzystajc z zasobw obliczeniowych w chmurze. Jest to szczeglnie przydatne dla analitykw danych, naukowcw zajmujcych si uczeniem maszynowym oraz programistw, ktrzy potrzebuj dostpu do potnych zasobw obliczeniowych bez koniecznoci posiadania odpowiedniego sprztu.

#### Gwne cechy Google Colab

1. **Darmowy dostp do zasobw obliczeniowych:** Google Colab oferuje bezpatny dostp do CPU, GPU i TPU. Uytkownicy mog korzysta z tych zasobw do trenowania modeli uczenia maszynowego oraz przeprowadzania oblicze wymagajcych duej mocy obliczeniowej.

2. **atwe udostpnianie i wsppraca:** Notatniki Colab mona atwo udostpnia innym uytkownikom za pomoc linkw, co umoliwia wspprac w czasie rzeczywistym. Kilka osb moe jednoczenie edytowa ten sam notatnik, co jest idealne dla zespow pracujcych nad wsplnymi projektami.

3. **Integracja z Google Drive:** Notatniki Colab s bezproblemowo zintegrowane z Google Drive, co umoliwia atwe przechowywanie, zarzdzanie i udostpnianie plikw. Uytkownicy mog rwnie bezporednio wczytywa i zapisywa dane w Google Drive.

4. **Bez instalacji i konfiguracji:** Google Colab dziaa w przegldarce, wic nie wymaga instalacji ani konfiguracji lokalnego rodowiska programistycznego. Wystarczy zalogowa si na konto Google, aby zacz korzysta z usugi.

5. **Obsuga bibliotek data science i uczenia maszynowego:** Colab ma preinstalowane popularne biblioteki Python, takie jak NumPy, Pandas, Matplotlib, Scikit-learn, TensorFlow i PyTorch, co umoliwia natychmiastowe rozpoczcie pracy nad projektami zwizanymi z analiz danych i uczeniem maszynowym.

# Kluczowe rnice ze rodowiskami developerskimi (PyCharm, Visual Studio Code)

### Kluczowe rnice ze rodowiskami developerskimi (PyCharm, Visual Studio Code)

Jupyter Notebook oraz tradycyjne rodowiska developerskie takie jak PyCharm i Visual Studio Code (VS Code) maj swoje unikalne cechy, ktre sprawiaj, e s bardziej odpowiednie dla rnych typw zada. Poniej omwione zostan kluczowe rnice midzy Jupyter Notebook a tymi rodowiskami developerskimi, co pomoe lepiej zrozumie, kiedy warto uywa kadego z tych narzdzi.

#### Jupyter Notebook

1. **Interaktywno:**
   - Jupyter Notebook jest narzdziem zaprojektowanym do interaktywnej pracy z kodem. Uytkownicy mog uruchamia kod fragmentami (komrkami) i natychmiastowo widzie wyniki, co jest idealne dla eksploracyjnej analizy danych, prototypowania algorytmw uczenia maszynowego oraz wizualizacji danych.
   - Komrki kodu i komrki Markdown umoliwiaj atwe mieszanie kodu z tekstem, co uatwia dokumentowanie i wyjanianie krokw analizy.

2. **atwo uycia:**
   - Jupyter Notebook jest bardzo intuicyjny i atwy w uyciu, zwaszcza dla pocztkujcych. Nie wymaga skomplikowanej konfiguracji i dziaa bezporednio w przegldarce.
   - Umoliwia szybkie rozpoczcie pracy z minimalnym czasem przygotowania rodowiska.

3. **Wizualizacja danych:**
   - Jupyter Notebook jest zoptymalizowany do wizualizacji danych. Wykresy i grafiki generowane przez biblioteki takie jak Matplotlib, Seaborn czy Plotly s wywietlane bezporednio w notatniku, co uatwia analiz wynikw.

4. **Eksperymentowanie i edukacja:**
   - Jest idealnym narzdziem do nauki i eksperymentowania, poniewa pozwala na szybkie testowanie i modyfikowanie fragmentw kodu. Czsto uywany jest w kursach online i warsztatach z zakresu data science i uczenia maszynowego.
   - Moliwo wsppracy w czasie rzeczywistym (np. w Google Colab) sprawia, e jest wietny do pracy zespoowej.

#### PyCharm

1. **Zaawansowane funkcje programistyczne:**
   - PyCharm, stworzony przez JetBrains, jest zaawansowanym zintegrowanym rodowiskiem programistycznym (IDE) dla jzyka Python. Oferuje funkcje takie jak automatyczne uzupenianie kodu, debugowanie oraz integracj z systemami kontroli wersji (np. Git).
   - Jest bardziej odpowiedni do pracy nad duymi projektami, gdzie zarzdzanie kodem, struktura projektu i zaawansowane narzdzia do debugowania s kluczowe.

2. **Wsparcie dla frameworkw:**
   - PyCharm wspiera wiele popularnych frameworkw webowych (Django, Flask) oraz narzdzi do testowania (pytest, unittest).. Posiada rwnie wbudowane wsparcie dla Docker i Kubernetes, co jest przydatne w tworzeniu aplikacji produkcyjnych.

3. **Profesjonalny rozwj oprogramowania:**
   - PyCharm jest narzdziem preferowanym przez profesjonalnych deweloperw pracujcych nad zoonymi aplikacjami. Jego zaawansowane funkcje pomagaj utrzyma wysok jako kodu oraz efektywnie zarzdza duymi bazami kodu.

#### Visual Studio Code (VS Code)

1. **Lekko i szybko:**
   - VS Code, rozwijany przez Microsoft, jest lekkim edytorem kodu, ktry obsuguje wiele jzykw programowania, w tym Python. Jest szybszy w uruchamianiu i mniej zasoboerny w porwnaniu do takich IDE jak PyCharm.

2. **Rozszerzalno:**
   - VS Code ma bogaty ekosystem rozszerze, ktre mog by instalowane, aby dostosowa rodowisko do potrzeb uytkownika. Istniej rozszerzenia do wsparcia Pythona, Jupyter Notebook, debugowania, integracji z Git i wielu innych funkcji.
   - Uytkownicy mog skonfigurowa VS Code jako lekkie IDE, dodajc tylko te funkcje, ktre s im potrzebne, co czyni go bardzo elastycznym narzdziem.

3. **Integracja z Jupyter Notebook:**
   - VS Code oferuje rozszerzenia, ktre umoliwiaj uruchamianie i edytowanie notatnikw Jupyter bezporednio w edytorze. To czy najlepsze cechy obu wiatw, umoliwiajc korzystanie z interaktywnoci Jupyter przy jednoczesnym zachowaniu zaawansowanych funkcji edytora kodu.

4. **Wsparcie dla wielu jzykw:**
   - VS Code jest zaprojektowany jako edytor wielojzykowy, co czyni go idealnym dla programistw pracujcych w rnych jzykach programowania. Dziki temu jest uniwersalnym narzdziem dla zespow pracujcych w zrnicowanych technologicznie rodowiskach.

### Podsumowanie

**Jupyter Notebook** jest narzdziem idealnym do interaktywnej analizy danych, prototypowania i edukacji, dziki swojej prostocie, interaktywnoci i wsparciu dla wizualizacji. 

**PyCharm** to zaawansowane IDE dla profesjonalnych deweloperw Pythona, oferujce peen zestaw narzdzi do tworzenia, testowania i utrzymania aplikacji na du skal.

**Visual Studio Code** czy zalety lekkiego edytora kodu z moliwoci rozbudowy o zaawansowane funkcje, oferujc wsparcie dla wielu jzykw i integracj z Jupyter Notebook.
