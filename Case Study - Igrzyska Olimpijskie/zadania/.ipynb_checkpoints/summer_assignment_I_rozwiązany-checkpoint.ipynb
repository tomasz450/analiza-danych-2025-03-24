### 2.

### 3.

Ile kobiet i ile mężczyzn wzięło udział w zmaganiach olimpijskich w poszczególnych latach?

```python
x = olimp.groupby(["Gender","Year"]).size().unstack(0)
x
```

### 4.

Utwórz ramkę z procentowym udziałem kobiet względem poszczególnych lat, w których odbywały się igrzyska.

```python
x["% Women"] = x.Women/(x.Men + x.Women) * 100
x["% Women"] = round(x["% Women"],1)
del x["Men"]
del x["Women"]
```

```python
x
```

### 5. 

Wyświetl ramkę 10 najbardziej utytułowanych medalistów zgodnie z rankingiem:
* brąz -- 1 punkt,
* srebro -- 2 punkty,
* złoto -- 6 punktów

```python
df = olimp.groupby(['Athlete', 'Medal']).size().unstack().fillna(0).astype('int')
df['Ranking'] = 6*df.Gold + 2*df.Silver + df.Bronze
df.sort_values(by='Ranking', ascending=False, inplace=True)
df = df['Gold Silver Bronze Ranking'.split()]
df.head(10)
```

### 6.

Przyjmując ranking z punktu 5. wyświetl 10 najbardziej utytułowanych medalistów (dowolnej płci) w dyscyplinie *Swimming*.

```python
df_2 = olimp.groupby('Discipline Athlete Medal'.split()).size()
df.head()
```

```python
swimming = df_2.loc['Swimming'].unstack().fillna(0).astype('int')
swimming.head()
```

```python
swimming['Ranking'] = 6*swimming.Gold + 2*swimming.Silver + swimming.Bronze
swimming.sort_values(by='Ranking', ascending=False, inplace=True)
swimming = swimming['Gold Silver Bronze Ranking'.split()]
swimming.head(10)
```
